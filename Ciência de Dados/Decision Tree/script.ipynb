```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
restaurante_df = pd.read_csv('restaurante.csv', delimiter=';', encoding='latin-1')
carro_df = pd.read_csv('carro.csv', delimiter=';', encoding='latin-1')
```

```python
def encode_categorical(df):
    le = LabelEncoder()

    for column in df.columns:
        if df[column].dtype == 'object':
            df[column] = le.fit_transform(df[column])

    return df
```

```python
restaurante_df = encode_categorical(restaurante_df)
carro_df = encode_categorical(carro_df)
```

```python
X_restaurante = restaurante_df.drop('R', axis=1)
y_restaurante = restaurante_df['R']

X_carro = carro_df.drop('Classe Social', axis=1)
y_carro = carro_df['Classe Social']
```

```python
dt_restaurante = DecisionTreeClassifier()
dt_restaurante.fit(X_restaurante, y_restaurante)

plt.figure(figsize=(12, 8))
plot_tree(dt_restaurante, feature_names=X_restaurante.columns, class_names=dt_restaurante.classes_.astype(str), filled=True)
plt.title('rvore de Deciso para Restaurantes')
plt.show()
```

```python
dt_carro = DecisionTreeClassifier()
dt_carro.fit(X_carro, y_carro)

plt.figure(figsize=(12, 8))
plot_tree(dt_carro, feature_names=X_carro.columns, class_names=dt_carro.classes_.astype(str), filled=True)
plt.title('rvore de Deciso para Carros')
plt.show()
```

# Anlise dos resultados:

- A rvore de deciso para o dataset de restaurantes mostra as condies de deciso com base nos atributos Experincia dos Profissionais (EP), Qualidade da Refeio (QR) e Localizao do Estabelecimento (LE) para prever se o restaurante ter Lucro ou Prejuzo. Esta rvore pode ser interpretada para entender quais fatores influenciam mais nas decises de negcio relacionadas a restaurantes.

- A rvore de deciso para o dataset de carros mostra as condies de deciso com base nos atributos Gnero, Tipo de Carro, Tamanho da Camisa e Classe Social do Proprietrio para prever a Classe Social do Proprietrio. Essa rvore pode ser usada para entender quais caractersticas dos proprietrios de carros esto mais relacionadas com sua classe social.
